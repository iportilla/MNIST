```python
from pathlib import Path
from ollama import chat

# from pathlib import Path

# Provide the correct path to the image
# path = Path('./samples/baby-bear.jpg')
# path = Path('../samples/jockey.jpg')
path = Path('../samples/dog-human.jpg')

# Ensure the file exists
if not path.exists():
    raise FileNotFoundError(f"Image not found at {path}")
```

```python
%%time
# Call the chat function
response = chat(
    # model='llama3.2-vision',
    model='gemma3',
    messages=[
        {
            'role': 'user',
            'content': 'What is in this image? Be concise.',
            'images': [str(path)],  # Convert Path object to string
        }
    ],
)
print(response.message.content)
# eval_duration=350760958
print(response.eval_duration)

print
```
